```python
c = ['First', 'Second']

fig = plt.figure(figsize=[40,25])
for i in range(0, 1):
    plt.subplot(1,1,i+1)
    plt.tick_params(labelsize=30)
#     item_df = bar5_exp_df[bar5_exp_df['品类'] == c[i]]
#     plt.bar(time_seq_list, item_df['浏览'], width=0.5)
    
#     for x, y in zip(time_seq_list,item_df['浏览']):
#         plt.text(x, y+0.3, y, ha='center', va='bottom', fontsize=40)
#     plt.legend(loc='upper right',labels=['浏览量'],fontsize = 40) if i==0 else 1
    
    ax = plt.gca()
    if i ==0:
        ax.set_yticks([0,72000])
        ax.set_yticklabels([],fontsize = 40)
    else:
        ax.set_yticks([0,30000])
        ax.set_yticklabels([],fontsize = 40)
    
#     ax.spines['right'].set_color('none')
#     ax.spines['left'].set_color('none')
#     ax.spines['top'].set_color('none')
#     plt.yticks([])
    
    # 不分组的情况下，click_df下标必须从 0开始
    click_df = bar5_exp_df[bar5_exp_df['品类'] == c[i]].reset_index(drop=True)
    
    ax1 = ax.twinx()
    ax1.plot(click_df['点击率'],'-',marker='*',linewidth =2,Markersize=10,color = 'r')
    plt.yticks([])
    plt.legend(loc='upper left',labels=['点击率'],fontsize = 40) if i==0 else 1
    if i ==0:
        ax1.set_yticks([0.015,0.02,0.03,0.04,0.05])
        ax1.set_yticklabels([],fontsize = 40)
    else:
        ax1.set_yticks([0.01,0.02,0.03])
        ax1.set_yticklabels([],fontsize = 40)
    
    plt.tick_params(labelsize=30)
#     plt.xticks([])
    for x, y in zip(time_seq_list, click_df['点击率']):
        ax1.text(x, y+0.0005, str(round(y*100,2))+'%', ha='center', va='bottom', fontsize=40, color = 'r')
        print("x,y: ", '--{}--'.format(x), '-{}-'.format(y))
#     plt.xticks(ticks=click_df['点击率'], labels=time_seq_list, step=2)
    ax1.spines['right'].set_color('none')
    ax1.spines['left'].set_color('none')
    ax1.spines['top'].set_color('none')
    plt.title('%s点击率'%c[i],fontsize = 50)
    
plt.show()

# ['<=1H', '(1H,6H]', '(6H,1D]', '(1D,3D]', '(3D,7D]', '>7D']
```
