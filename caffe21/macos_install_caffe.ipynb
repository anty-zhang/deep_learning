# caffe 依赖包下载与编译


##### **ProtoBuffer**

主页：https://github.com/google/protobuf/releases?after=v2.6.1
ProtoBuffer是由Goole开发的一种可以实现内存与非易失存储介质（比如硬盘文件）交换的协议接口，支持C++／Java／Python。在caffe中大量使用ProtoBuffer作为权值和模型参数的载体。

```bash
$wget https://github.com/google/protobuf/releases/download/v2.5.0/protobuf-2.5.0.tar.gz
$tar -zxvf protobuf-2.5.0.tar.gz
$cd protobuf-2.5.0
$./configure --prefix=/data/dp/caffe/caffe_install/
$make
$make install

```


##### **Boost**

主页： http://www.boost.org/users/history/version_1_56_0.html

“C++标准库”，是一个功能强大、构造精巧、跨平台、开源且免费的库。使用了很多现代编程技术，内容涵盖字符串处理、正则、容器、数据接口、并发开发、函数时编程、范型编程、设计模式等。

```sh
$wget http://sourceforge.net/projects/boost/files/boost/1.56.0/boost_1_56_0.tar.bz2
$tar jxvf boost_1_56_0.tar.bz2
$cd cd boost_1_56_0
$./bootstrap.sh --with-libraries=system,thread,python
$cp -r boost /data/dp/caffe/caffe_install/include/
$cp stage/lib/* /data/dp/caffe/caffe_install/lib/

```

##### ** BLAS **

主页：https://github.com/xianyi/OpenBLAS/releases

BLAS（Basic Linear AlgeBra Subprograms，基本线性代数子程序），主要用于数学计算，最常用的BLAS实现有 **Intel MKL、ATLAS、OpenBLAS **。caffe可以选择其中任意种。在GPU端的数值计算则有对应的cuBLAS完成。

Intel数学核心函数库（MKL）是一套高度优化、线程安全的数学例程、函数，面向高性能的工程、科学与财务应用。英特尔 MKL 的集群版本包括 ScaLAPACK 与分布式内存快速傅立叶转换，并提供了线性代数 (BLAS、LAPACK 和Sparse Solver)、快速傅立叶转换、矢量数学 (Vector Math) 与随机号码生成器支持。

主要包括：

① LAPACK (线形代数工具linear algebra package)

② DFTs (离散傅立叶变换 Discrete Fourier transforms)

③ VML (矢量数学库Vector Math Library)

④ VSL (矢量统计库Vector Statistical Library)


ATLAS是python下的一个线性代数库，是基于另外两个线性代数库BLAS和lapack的；

NumPy提供了一个在python中做科学计算的基础库，它重在数值计算，甚至可以说是用于多维数组处理的库；

SciPy是基于numpy，提供了一个在python中做科学计算的工具集，也就是说它是更上一个层次的库；

Theano则是基于NumPy以及SciPy的一个更高级的用于科学计算的库。



```sh
$wget https://github.com/xianyi/OpenBLAS/archive/v0.2.14.tar.gz -O OpenBLAS-0.2.14.tar.gz
$tar -zxvf OpenBLAS-0.2.14.tar.gz 
$cd OpenBLAS-0.2.14
$make -j

```

##### ** HDF5 **

主页： https://support.hdfgroup.org/ftp/HDF5/releases/hdf5-1.8.9/src/

HDF（Hierarchical Data File）是美国国家高级计算机应用中心（NCSA）为了满足各种领域的研究需求而研制的一种高效存储和分发科学数据的新型数据格式。 它可以存储不同类型的图像和数码数据的文件，并且可以在不同的机器上传输，同时还有统一处理这种格式的函数库。

```sh
$wget https://support.hdfgroup.org/ftp/HDF5/releases/hdf5-1.8.9/src/hdf5-1.8.9.tar.gz
$tar -zxvf hdf5-1.8.9.tar.gz 
$cd hdf5-1.8.9
$./configure --prefix=/data/dp/caffe/caffe_install/
$make -j
$make install
```

##### ** LMDB/LEVELDB **

主页：http://www.linuxfromscratch.org/blfs/view/svn/server/lmdb.html
LMDB（Lightning Memory-Mapped Database Mananger）----闪电般的内存映射型数据库管理器。在caffe中的主要作用是提供数据管理，例如，将JPEG，二进制数据转换为统一的Key-Value存储，便于Caffe的DataLayer获取这些数据。
LEVELDB，是caffe早起版本使用的数据存储方式，由Google开发。是一种持续的键值对存储方式，键和值可以是任意字节数组，键的存储顺序可由用户定义的比较函数决定，目前大部分caffe例程都已经用LMDB替代了LEVELDM。

```sh
$wget http://pkgs.fedoraproject.org/repo/pkgs/leveldb/leveldb-1.7.0.tar.gz/5c3c6cfbb16a21cdbea0db656aa32fec/leveldb-1.7.0.tar.gz
$tar -zxvf leveldb-1.7.0.tar.gz
$cd leveldb-1.7.0 && make
$cp -r include/leveldb /data/dp/caffe/caffe_install/include/
$cp libleveldb.* /data/dp/caffe/caffe_install/lib/

$wget  https://github.com/LMDB/lmdb/archive/LMDB_0.9.19.tar.gz
$tar -zxvf LMDB_0.9.19.tar.gz
$cd lmdb-LMDB_0.9.19/libraries/liblmdb
$make
$cp *.so /data/dp/caffe/caffe_install/lib/
$cp lmdb.h /data/dp/caffe/caffe_install/include/

```



##### **Snappy **

Snappy是一个用来压缩和解压缩的C++库。 比zlib快，但文件相对要大20%～100%。

```sh
$wget http://pkgs.fedoraproject.org/repo/pkgs/snappy/snappy-1.1.1.tar.gz/8887e3b7253b22a31f5486bca3cbc1c2/snappy-1.1.1.tar.gz
$tar -zxvf snappy-1.1.1.tar.gz 
$cd snappy-1.1.1
$ls
$./configure  --prefix=/data/dp/caffe/caffe_install/
$make
$make install

```


# 使用brew安装

```sh
brew install -vd snappy
ll /usr/local/Homebrew/Library/Taps/homebrew/homebrew-core/Formula/snappy.rb
ll /usr/local/Homebrew/Library/Homebrew
brew install -vd leveldb lmdb >> /data/dp/caffe/caffe_install/install.md
brew install -vd gflags glog >> /data/dp/caffe/caffe_install/install.md
brew install -vd szip >> /data/dp/caffe/caffe_install/install.md
brew tap homebrew/science
brew install -vd hdf5 opencv >> /data/dp/caffe/caffe_install/install.md 
brew install -vd protobuf boost >> /data/dp/caffe/caffe_install/install.md 

# openblas dependencies gmp, mpfr, libmpc, isl, gcc
brew install --fresh -vd openblas

BLAS_INCLUDE := /usr/local/opt/openblas/include
BLAS_LIB := /usr/local/opt/openblas/lib
git clone https://github.com/BVLC/caffe.git
make -j

```
